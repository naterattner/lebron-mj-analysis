# Comparing LeBron James and Michael Jordan over 15 NBA Seasons
Plotting nine key statistical categories by season and by age. Data source: https://www.basketball-reference.com. A blog post breaking down the below data and charts can be [found here](https://medium.com/@naterattner/charting-15-years-of-michael-jordan-and-lebron-james-1ae1ccbd1c7d).

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

Load CSV

```python
df = pd.read_csv("lebron-mj.csv")
```

```python
lebron_mj_all = df
lebron_mj_all
```

```python
lebron_age = lebron_mj_all.head(15)
mj_age = lebron_mj_all.tail(19)
```

Remove seasons MJ did not play

```python
seasons_played = df.drop([24, 29,30,31])
seasons_played
```

```python
lebron = seasons_played.head(15)
mj = seasons_played.tail(15)
```

```python
# Get rid of error messaging
pd.options.mode.chained_assignment = None  # default='warn'

#Add season_number to dataframe
lebron['Season Number'] = pd.Series(range(1,16))
lebron
```

```python
#Add season_number to dataframe
mj = mj.reset_index(drop=True)
mj['Season Number'] = pd.Series(range(1,16))
mj
```

```python
# One subplot
fig, ax = plt.subplots()

# Use ax for both
lebron.plot(x='Season Number', y='PTS', ax=ax, label='LeBron James', color="navy")
mj.plot(x='Season Number', y='PTS', ax=ax, label='Michael Jordan', color="red")
ax.set_title("Points per Game")
```

Subplots by Season Number

```python
# Subplots with three rows of three columns each
# Using figsize to make the figure 20"x10"
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(nrows=3, ncols=3, figsize=(20,10))

# PTS
lebron.plot(x='Season Number', y='PTS', ax=ax1, label='LeBron James', color="navy")
mj.plot(x='Season Number', y='PTS', ax=ax1, label='Michael Jordan', color="red")
ax1.set_title("Points", fontsize=20)
ax1.set_xticks(range(1, 16, 2))
ax1.set_yticks(range(20, 38, 4))
ax1.tick_params(labelsize=12)
ax1.set_xlabel("Season Number", fontsize=15)
ax1.legend(fontsize='15', bbox_to_anchor=(-.025, 1.85), loc=2,)

#TRB
lebron.plot(x='Season Number', y='TRB', ax=ax2, legend=False, color="navy")
mj.plot(x='Season Number', y='TRB', ax=ax2, legend=False, color="red")
ax2.set_title("Rebounds", fontsize=20)
ax2.set_xticks(range(1, 16, 2))
ax2.tick_params(labelsize=12)
ax2.set_xlabel("Season Number", fontsize=15)

#AST
lebron.plot(x='Season Number', y='AST', ax=ax3, legend=False, color="navy")
mj.plot(x='Season Number', y='AST', ax=ax3, legend=False, color="red")
ax3.set_title("Assists", fontsize=20)
ax3.set_xticks(range(1, 16, 2))
ax3.tick_params(labelsize=12)
ax3.set_xlabel("Season Number", fontsize=15)

#BLK
lebron.plot(x='Season Number', y='BLK', ax=ax4, legend=False, color="navy")
mj.plot(x='Season Number', y='BLK', ax=ax4, legend=False, color="red")
ax4.set_title("Blocks", fontsize=20)
ax4.set_xticks(range(1, 16, 2))
ax4.tick_params(labelsize=12)
ax4.set_xlabel("Season Number", fontsize=15)

#STL
lebron.plot(x='Season Number', y='STL', ax=ax5, legend=False, color="navy")
mj.plot(x='Season Number', y='STL', ax=ax5, legend=False, color="red")
ax5.set_title("Steals", fontsize=20)
ax5.set_xticks(range(1, 16, 2))
ax5.tick_params(labelsize=12)
ax5.set_xlabel("Season Number", fontsize=15)

#TOV
lebron.plot(x='Season Number', y='TOV', ax=ax6, legend=False, color="navy")
mj.plot(x='Season Number', y='TOV', ax=ax6, legend=False, color="red")
ax6.set_title("Turnovers", fontsize=20)
ax6.set_xticks(range(1, 16, 2))
ax6.tick_params(labelsize=12)
ax6.set_xlabel("Season Number", fontsize=15)

#FG%
lebron.plot(x='Season Number', y='FG%', ax=ax7, legend=False, color="navy")
mj.plot(x='Season Number', y='FG%', ax=ax7, legend=False, color="red")
ax7.set_title("Field Goal PCT", fontsize=20)
ax7.set_xticks(range(1, 16, 2))
ax7.tick_params(labelsize=12)
ax7.set_xlabel("Season Number", fontsize=15)

#3P%
lebron.plot(x='Season Number', y='3P%', ax=ax8, legend=False, color="navy")
mj.plot(x='Season Number', y='3P%', ax=ax8, legend=False, color="red")
ax8.set_title("Three Point PCT", fontsize=20)
ax8.set_xticks(range(1, 16, 2))
ax8.tick_params(labelsize=12)
ax8.set_xlabel("Season Number", fontsize=15)

#MP
lebron.plot(x='Season Number', y='MP', ax=ax9, legend=False, color="navy")
mj.plot(x='Season Number', y='MP', ax=ax9, legend=False, color="red")
ax9.set_title("Minutes Played", fontsize=20)
ax9.set_xticks(range(1, 16, 2))
ax9.tick_params(labelsize=12)
ax9.set_xlabel("Season Number", fontsize=15)


# Tight layout to avoid overlap
plt.tight_layout()

#Titles and spacing
plt.suptitle('LeBron James vs. Michael Jordan', fontsize=26, weight="bold")
plt.subplots_adjust(hspace=0.6, wspace=0.15, top=0.83)
plt.text(x = -23.5, y = 113, s = "Statistics by season over the course of 15 years in the NBA",
               fontsize = 24)

#Bottom bar
plt.text(x = -36.5, y = 14, s = '__________________________________________________________________________________________________________________________________________________________________________________________________________________________________________________________________________________________________',
    color = 'grey', alpha = .7)
plt.text(x = -36.5, y = 11, s = 'Source: Basketball-Reference.com                                                                                                                                                                             ',
    fontsize=16, color = 'grey', alpha = .7)

#Save chart as png
#plt.savefig('lebron_mj_seasons.png')
```
